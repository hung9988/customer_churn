```python
import numpy as np
import matplotlib.pyplot as plt
import argparse
import pandas as pd


raw_data=pd.read_csv('train.csv')
def one_hot(X):
    features_list=[]
    m,n=X.shape
    X_new = np.empty((m,0))
    index=0
    for i in X[0,:]:
        if isinstance(i,str):
            uniques, ids= np.unique(X[:,index], return_inverse=True)
            one_hot_temp = np.zeros((m, len(uniques)))
            one_hot_temp[np.arange(m), ids] = 1
            features_list.append(one_hot_temp)
            X_new=np.concatenate((X_new,one_hot_temp),axis=1)
        else:
            features_list.append(X[:,index].reshape(-1,1))
            X_new= np.concatenate((X_new,X[:,index].reshape(-1,1)),axis=1)
        index+=1
    return X_new,features_list

data=np.array(raw_data)
data[data=='no']=0
data[data=='yes']=1
data, features_list=one_hot(data)

data = data.astype(float)

def normalize(data,features):
    for i in features:
        data[:,i] = (data[:,i] - data[:,i].mean()) / data[:,i].std()
    
    return data

data = normalize(data,np.arange(0,data.shape[1]-1))




def train_valid_test_split(data, split_train,split_valid,split_test):
    np.random.shuffle(data)
    split_train=int(split_train*len(data))
    split_valid=int(split_valid*len(data))
    split_test=int(split_test*len(data))
    return data[:split_train], data[split_train:split_train+split_valid], data[split_train+split_valid:split_train+split_valid+split_test]
    

train_data, valid_data, test_data = train_valid_test_split(data, 0.6, 0.2, 0.2)

X_train, y_train = train_data[:,0:-1], train_data[:,-1]
X_valid, y_valid = valid_data[:,0:-1], valid_data[:,-1]
X_test, y_test = test_data[:,0:-1], test_data[:,-1]


y_train=y_train.reshape(-1,1)
y_valid=y_valid.reshape(-1,1)
y_test=y_test.reshape(-1,1)

np.savez('churn_data_full.npz', X_train=X_train, y_train=y_train, X_valid=X_valid, y_valid=y_valid, X_test=X_test, y_test=y_test)
```
