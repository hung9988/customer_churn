```python
from PIL import Image
```

```python
img=Image.open(r'cat.jpg')
```

```python
import numpy as np

img_array = np.array(img)
combined_img = np.mean(img_array, axis=2)

```

```python
import matplotlib.pyplot as plt
plt.imshow(combined_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
from scipy.ndimage import convolve

# Define the edge detection filter
filter = np.array([[1, 1, 1],
                   [1, -10, 1]])

# Apply the filter to the image
edges = convolve(combined_img, filter)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()
```

```python
from scipy.ndimage import maximum_filter

# Apply max pooling to the edges
pooled_edges = maximum_filter(edges, size=3)

# Display the pooled edges
plt.imshow(pooled_edges, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import numpy as np

# Define the pooling function
def max_pooling(image, pool_size=(2, 2)):
    height, width = image.shape
    pool_height, pool_width = pool_size

    # Calculate the output dimensions
    output_height = height // pool_height
    output_width = width // pool_width

    # Initialize the pooled image
    pooled_image = np.zeros((output_height, output_width))

    # Perform max pooling
    for i in range(output_height):
        for j in range(output_width):
            # Extract the region of interest
            roi = image[i * pool_height : (i + 1) * pool_height, j * pool_width : (j + 1) * pool_width]

            # Find the maximum value in the region
            pooled_value = np.max(roi)

            # Assign the maximum value to the pooled image
            pooled_image[i, j] = pooled_value

    return pooled_image

# Apply max pooling to the image
pooled_image = max_pooling(edges)

# Print the dimensions of the pooled image
print(pooled_image.shape)
```
